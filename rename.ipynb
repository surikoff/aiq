```python
import os
from urllib.parse import quote

files_list = [each.split("-predict")[0] for each in os.listdir("emotional_dialogues") if each.endswith('-predict.txt')]
files_list
```

```python

import glob
import json
from os import path, listdir, scandir
from yattag import Doc
from random import randint
from transliterate import slugify

OUTPUT_FILE = "character_tests.html"


doc, tag, text = Doc().tagtext()

with tag('html'):
    with tag('body'):
        for file in files_list:
                file_name = slugify(file)
                os.rename(f"emotional_dialogues/{file}.txt", f"emotional_dialogues/{file_name}.txt")
                os.rename(f"emotional_dialogues/{file}-predict.txt", f"emotional_dialogues/{file_name}-predict.txt")
                with tag('p'):
                    with tag('a', target="_blank", href=f"{file_name}.txt"):
                        text(file)
                    text(" => ")
                    with tag('a', target="_blank", href=f"{file_name}-predict.txt"):
                        text("predict")
                    doc.stag("hr")

with open("emotional_dialogues/index.html", "w", encoding="utf-8") as f:
    f.write(doc.getvalue())
```
